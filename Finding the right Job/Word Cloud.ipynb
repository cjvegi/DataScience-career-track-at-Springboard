## Job Responsibilities and Skill set

***In this notebook, we will use job descriptions and skills required for job titles such as Data Analyst, Data scientist, Data Engineer and their variants from Linkedin, kaggle and glassdoor. Using this information, we will generate a word cloud to visualize what skills are responsibilities are required for job title of interest***

```python
from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt


## Data analyst responsibilities
f = open('data/Data_analyst_responsibility.txt','r')
data_analyst_resp = f.read()
print('*'* 100)
print(''* 100)
print(data_analyst_resp)
f.close()



##### Data analyst skills
f = open('data/Data_analyst_skill.txt','r')
data_analyst_skill = f.read()
print('*'* 100)
print(''* 100)
print(data_analyst_skill)
f.close()



##### Data scientist responsibilities
f = open('data/data_scientist_responisibilities.txt','r')
data_scientist_responisibilities = f.read()
print('*'* 100)
print(''* 100)
print(data_scientist_responisibilities)
f.close()

##### Data scientist skills
f = open('data/data_scientist_skills.txt','r')
data_scientist_skills = f.read()
print('*'* 100)
print(''* 100)
print(data_scientist_skills)
f.close()
```

```python
def word_cloud_job_title(data, font_size = 40, title = '') :
    
    stopwords = ['etc','years', 'Etc','degree','skill','using','preferred','field',
                'based','related','including','Ability']
    for word in stopwords:
        if word in data:
            data=data.replace(word,"")
        
    # Generate a word cloud image
    wordcloud = WordCloud().generate(data)
    
    # Display the generated image:
    # the matplotlib way:
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    fig = plt.gcf()
    fig.set_size_inches(15,10)
    plt.title(title, fontsize = 24)
    
    # lower max_font_size
    wordcloud = WordCloud(max_font_size=font_size).generate(data)
    plt.figure()
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    fig = plt.gcf()
    fig.set_size_inches(15,10)
    plt.title(title, fontsize = 24)
    plt.show()
```

```python
### Data_analyst responsibilites
word_cloud_job_title(data_analyst_resp, title = 'data_analyst_responsibility')
```

```python
### Data_analyst skill
word_cloud_job_title(data_analyst_skill, title = 'data_analyst_skill')
```

```python
### Data scientist responsibilites
word_cloud_job_title(data_scientist_responisibilities, title = 'data_scientist_responisibilities')
```

```python
### Data scientist skills
word_cloud_job_title(data_scientist_skills, title='data_scientist_skills')
```

***We notice that, data analyst roles require woorking in team and developing some data products and requires the candidate to have mathematical, statistical, SQL background. Data scientist is more directly responsibility to help develop predictive models, support business while working in a team. Data scientist also requires more skills set ranging from programming langugaes like python, machine learning , data bases and have strong quantitative background***
